# Field Enhancement at Semi-cylinder
Planar electrostatics on a parallel-plate structure with a semi-cylindrical protrusion. This causes a field enhancement $f= 2$ at the top of the protrusion, and is used to model e.g. welding joints in high-voltage technology.

## Configuration
The configuration consists of a parallel-plate structure with width $W$ and height $H$, with a semi-cylindrical protrusion of radius $r$ at the bottom electrode. A voltage of $U_0$ is applied between the electrodes to produce a background field $E_0 = \frac{U_0}{H}$.
![Configuration semicylinder](figures/setup_semicylinder.svg)

## Theory
Simple planar electrostatics without space charge, solving the Laplace equation:
$$ \nabla^2 V = 0 $$
In weak form this becomes
$$ \int_\Omega \nabla u \cdot \nabla v d\Omega = 0 $$
where it was assumed that all the Neumann boundary conditions are symmetry conditions (on the two sides of the simulation domain).

```julia
using LinearAlgebra

using Plots
using LaTeXStrings

using gmsh

using Gridap
using GridapGmsh
```

# Create Geometry

```julia
W = 0.3;     # Width of the simulation domain [m]
H = 1;       # Height of the simulation domain [m]
r = 1e-2;    # Radius of cylinder [m]

E0 = 1e5;    # Background field [V/m]
U0 = E0 * H; # Required voltage to create background field [V]
```

```julia
gmsh.finalize()
gmsh.initialize()
gmsh.option.setNumber("General.Terminal", 1)
gmsh.model.add("fieldenh_semicylinder")

c1 = W / 10;
c2 = r / 100;
c3 = r / 10;

# Points
gmsh.model.geo.addPoint(-W/2, 0, 0, c1, 1)
gmsh.model.geo.addPoint(-r/2, 0, 0, c2, 2)
gmsh.model.geo.addPoint(0, 0, 0, c3, 3)
gmsh.model.geo.addPoint(r/2, 0, 0, c2, 4)
gmsh.model.geo.addPoint(W/2, 0, 0, c1, 5)
gmsh.model.geo.addPoint(W/2, H, 0, c1, 6)
gmsh.model.geo.addPoint(-W/2, H, 0, c1, 7)

# Lines
gmsh.model.geo.addLine(1, 2, 1)
gmsh.model.geo.addLine(2, 3, 2)
gmsh.model.geo.addLine(3, 4, 3)
gmsh.model.geo.addLine(4, 5, 4)
gmsh.model.geo.addLine(5, 6, 5)
gmsh.model.geo.addLine(6, 7, 6)
gmsh.model.geo.addLine(7, 1, 7)

gmsh.model.geo.addCircleArc(4, 3, 2, 8)

# Surfaces
gmsh.model.geo.addCurveLoop([1, -8, 4, 5, 6, 7], 1)
gmsh.model.geo.addCurveLoop([2, 3, 8], 2)

gmsh.model.geo.addPlaneSurface([1], 1)
gmsh.model.geo.addPlaneSurface([2], 2)

# Physics
gmsh.model.addPhysicalGroup(2, [1], 1)
gmsh.model.addPhysicalGroup(2, [2], 2)

gmsh.model.addPhysicalGroup(1, [1, 2, 3, 4, 8], 1)
gmsh.model.setPhysicalName(1, 1, "D1")
gmsh.model.addPhysicalGroup(0, [1, 2, 3, 4, 5], 1)
gmsh.model.setPhysicalName(0, 1, "D1p")

gmsh.model.addPhysicalGroup(1, [6], 2)
gmsh.model.setPhysicalName(1, 2, "D2")
gmsh.model.addPhysicalGroup(0, [6, 7], 2)
gmsh.model.setPhysicalName(0, 2, "D2p")

gmsh.model.addPhysicalGroup(1, [5, 7], 3)
gmsh.model.setPhysicalName(1, 3, "N1")

# Generate mesh and save
gmsh.model.geo.synchronize()
gmsh.model.mesh.generate(2)

gmsh.write("fieldenh_semicylinder.msh")
```

```julia
gmsh.fltk.run()
```

# Gridap

```julia
# Load geometry from gmsh
model = GmshDiscreteModel("fieldenh_semicylinder.msh")

# Create a reference finite element space to use for the construction of test space V0
#  The FE space is Lagrangian and of order 1
order = 2
reffe = ReferenceFE(lagrangian, Float64, order);
# The test space is created from the reference FE with
#  conformity = H1: shape functions are continuous
#  D1 and D2 being the boundary points with Dirichlet condition applied
V0 = TestFESpace(model, reffe; conformity = :H1, dirichlet_tags = ["D1", "D1p", "D2", "D2p"]);

# Trial space is constructed from the test space
#  the [0, 1] indicates the values of the Dirichlet condition at D1 and D2
Ug = TrialFESpace(V0, [0, 0, U0, U0]);

# Triangulation of the domain
#  degree must be at least 2 times the order of the reference space
degree = 2 * order;
Ω = Triangulation(model);
dΩ = Measure(Ω, degree);
```

```julia
# Construct weak form
lhs(u,v) = ∫( ∇(v) ⋅ ∇(u) )dΩ;
rhs(v)   = 0;

# Construct FE operator with LHS, RHS, and test and trial spaces
op = AffineFEOperator(lhs, rhs, Ug, V0);

# Solve the linear FE system with LU solver
ls = LUSolver()
solver = LinearFESolver(ls)

uh = solve(solver, op);
E = ∇(uh);
```

# Plot Field

## Along $y$ axis

```julia
y = Vector(0:1e-4:5e-2)
coord = [Point(0, y[i]) for i in 1:length(y)];
uline = uh(coord);
Eline = E(coord);
Eline = [sqrt(E[1]^2 + E[2]^2)  for E in Eline];
```

```julia
plot(y * 1e2, Eline * 1e-5, color = :black, legend = :false, xlabel = L"y\,\mathrm{[cm]}", ylabel = L"|E|\,\mathrm{[kV/cm]}")
```

```julia
f = maximum(Eline) / E0
```

As expected, the field enhancement factor $f \approx 2$

## Along $x$ axis

```julia
x = Vector(-5*r:1e-4:5*r)
coord = [Point(x[i], 0) for i in 1:length(x)];
uline = uh(coord);
Eline = E(coord);
Eline = [sqrt(E[1]^2 + E[2]^2)  for E in Eline];
```

```julia
plot(x * 1e2, Eline * 1e-5, color = :black, legend = :false, xlabel = L"x\,\mathrm{[cm]}", ylabel = L"|E|\,\mathrm{[kV/cm]}")
```

In the corners, the field becomes zero

# Export to VTK

```julia
writevtk(Ω,"fieldenh_semicylinder", cellfields=["uh" => uh, "E" => E])
```
